$x = [x1, x2,  xn] $ is a 1 column vector where $x_1x_n$ are vectors containing each the m observation. essentially a matrix expressed in vector formed.

### Each PC is decorelated to the preceding one. 
Because PCs are orthogonal. Remember that orthogonal vectors means that their dot product is equal to 0. You can choose to keep only the first few PCs, knowing that each PC is a linear combination of the data features. For instance, one PC could be a linear combination of the length of the muscial piece and the number of instruments playing.

https://hadrienj.github.io/posts/Deep-Learning-Book-Series-2.12-Example-Principal-Components-Analysis/

The encoding function f(x) transforms x into c and the decoding function transforms back c into an approximation of x. 

### Decoding function

Decoding function: 
$ g(c) = Dc $
- matrix $D$ must be orthogonal 
- $D$ must have unit norm 
- let dimension of be $n*l = (n*l)(l*l) $ for $ g(c) = Dc $

### Encoding function

we want to miniminze distance between $x$ and $g(c)$ thus we use $L^2$ norm

$$ || x - g(c) ||^2_2 $$

we minimize it:
$$c*  = argmin|| x - g(c) ||^2_2  $$

we can sub $L_2$ norm with $||y||^2_2 = y^Ty  $

$$ 
\begin{align*} 
(x- g(c))^T(x-g(c)) & = (x^T- g(c)^T)(x-g(c)) \\
&= x^Tx -x^Tg(c) -g(c)^Tx + g(c)^Tg(c)  \tag{expand}
\end{align*} 
$$

since vector mulitplication is commutative $x^Ty = y^Tx$ , sub $ g(c)^T = x^Tg(c) $ thus

$$ x^Tx - 2x^Tg(c) + g(c)^Tg(c) $$

remove $ x^Tx $ since indepedent from c when minimized
$$ c* = argmin -2x^Tg(c) + g(c)^Tg(c)   $$

sub g(c) = Dc

$$  
\begin{align*} 
c* &= argmin -2x^TDc + (Dc)^TDc \\
&= argmin  -2x^TDc + c^TD^TDc \tag{D^TD = I since orthogonal D} \\
& = argmin -2x^TDc + c^Tc
\end{align*} 
$$

#### Find gradient for encoding function 

$\nabla_c (c*) =  0  $

1st term
$$ \frac{d(-2x^TDc)}{dc} = -2x^TD $$

2nd term, by expansion
$$  \frac{d(c^Tc)}{dc} =  2c $$

thus $ -2x^TD + 2c = 0 $

$c =x^TD $

since we want $c$ to be column vector (l, 1) we transpose 

Thus encoding function $ c= (x^TD)^T  = D^Tx$

To decode, we use $g(c) = Dc $ thus:
    
$ r(x) = g(f(x)) = D(D^Tx) =DD^Tx  $

where r(x) is reconstruction of 1 data point

### Find D

purpose of the PCA is to change the coordinate system in order to maximize the variance along the first dimensions of the projected space. This is equivalent to minimizing the error between data points and their reconstruction.

using frobenius norm

$$ D* = argmin \sqrt{\Sigma_{i,j} (x_j^i - r(x^i)_j )^2} $$

find only the 1st PC, thus l = 1, and thus D.shape = (n, 1), which we can call d since it is vector

$$ d^* = argmin \Sigma_{i} || { (x^i - r(x^i) )} ||^2_2 $$
 
sub $ r(x) = dd^Tx  $

$$ d^* = argmin \Sigma_{i} || { (x^i - dd^Tx^i )} ||^2_2 $$

tranpose to get $x^{(i)T}$ instead of $x^i$.

we want to transpose to get the rows of $X$ instead of col
$$
\begin{align*} 
d^* &= argmin \Sigma_{i} || (x^i - dd^Tx^i )^T ||^2_2  \\
& = argmin \Sigma_{i} || x^{(i)T} - x^{(i)T}dd^T  ||^2_2 \\
    & = argmin || X- Xdd^T  ||^2_F  \tag{d^Td = 1}
\end{align*} 
$$

since frobenius norm can be expressed via trace matrix $ A_F= \sqrt{ Tr( AA^T)  }$

$$ d^* = argmin Tr  ((X - Xdd^T )(X- Xdd^T)^T )  $$

swap order of 2 brackets of matrices (as trace allows use to)

$$ d^* = argmin Tr  ((X^T - (Xdd^T)^T)(X - Xdd^T ) )  $$

since expanding $(Xdd^T )^T = dd^TX^T  $

$$
\begin{align*} 
d^* & = argmin Tr  ((X^T - dd^TX^T)(X - Xdd^T ) )  \\
& = argmin Tr  ((X^TX  - X^TXdd^T - dd^TX^TX + dd^TX^TXdd^T ) \\
& = argmin Tr  (X^TX)  -  Tr(X^TXdd^T) - Tr(dd^TX^TX) + Tr(dd^TX^TXdd^T) \\
& = argmin -Tr(X^TXdd^T) - Tr(dd^TX^TX) + Tr(dd^TX^TXdd^T) \tag{remove 1st term as no d} \\
& = argmin -2Tr(X^TXdd^T) + Tr(X^TXdd^Tdd^T) \tag{can rearrange matrix order} \\
& = argmin -Tr(d^TX^TXd) \tag{d^Td = 1} \\
& = argmax Tr(d^TX^TXd) \tag{swap sign}
\end{align*}  
$$





#### we center our data around 0 so that $X^TX $ is the covariance matrix. Why center? 

If $\mu_x = 0$, $Cov(x) = \Sigma^n_{i=0} (x_i - mu_x)(xi - mu_x)^T $ becomes $Cov(x) = \Sigma^n_{i=0} x_ixi^T $ thus $ C = XX^T $

The covariance matrix is a n by n matrix (n being the number of dimensions). Its diagonal is the variance of the corresponding dimensions and the other cells are the covariance between the two corresponding dimensions (the amount of redundancy). To maximize the variance and minimize the covariance (in order to decorrelate the dimensions), the ideal covariance matrix is a diagonal matrix (non-zero values in the diagonal only). Therefore the diagonalization of the covariance matrix will give us the optimal solution.

Also centering means PCA regression model is without intercept (https://stats.stackexchange.com/questions/22329/how-does-centering-the-data-get-rid-of-the-intercept-in-regression-and-pca)

```python
# create 2D dataset 

import numpy as np

np.random.seed(123)
x = 5*np.random.rand(100)
y = 2*x +1 + np.random.randn(100)

x = x.reshape(100, 1)
y = y.reshape(100, 1)

X = np.hstack((x, y))
X.shape
```

```python
import matplotlib.pyplot as plt

plt.plot(X[:,0], X[:,1], '*')
plt.show()
```

```python
def centerData(X):
    X = X.copy()
    X -= np.mean(X, axis =0) # minus off mean 
    return X
```

```python
X_centered = centerData(X)
plt.plot(X_centered[:,0], X_centered[:,1], '*')
plt.show()
```

to calculate $ d* = argmax Tr(d^TX^TXd) $ subject to $ d^Td$ ,  we find eigenvectors $ X^TX$ first 

```python
eigVals, eigVecs = np.linalg.eig(X_centered.T @ X_centered)
eigVecs
```

```python
eigVals
```

```python
# make plotting function

def plotVectors(vecs, cols, alpha=1):
    """
    Plot set of vectors.

    Parameters
    ----------
    vecs : array-like
        Coordinates of the vectors to plot. Each vectors is in an array. For
        instance: [[1, 3], [2, 2]] can be used to plot 2 vectors.
    cols : array-like
        Colors of the vectors. For instance: ['red', 'blue'] will display the
        first vector in red and the second in blue.
    alpha : float
        Opacity of vectors

    Returns:

    fig : instance of matplotlib.figure.Figure
        The figure of the vectors
    """
    plt.figure()
    plt.axvline(x=0, color='#A9A9A9', zorder=0)
    plt.axhline(y=0, color='#A9A9A9', zorder=0)

    for i in range(len(vecs)):
        x = np.concatenate([[0,0],vecs[i]])
        plt.quiver([x[0]],
                   [x[1]],
                   [x[2]],
                   [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=cols[i],
                   alpha=alpha)
```

```python
orange = '#FF9A13'
blue = '#1190FF'
plotVectors(eigVecs.T, [orange, blue])
plt.plot(X_centered[:,0], X_centered[:,1], '*')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.show()
```

vector associated with the larger eigenvalue tells us the direction associated with the larger variance in our data (blue line)

since the encoding function is $c= D^Tx $ ,  $D$ is the eigenvector matrix we calculated. We apply it to all data points, and since $X$ has dimensions on the columns we need to transpose it

```python
X_new = eigVecs.T @ X_centered.T

# formula c= D^Tx is for each data point. we apply formula to each data point on the graph

plt.plot(eigVecs.T.dot(X_centered.T)[0, :], eigVecs.T.dot(X_centered.T)[1, :], '*')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.show()
```

The rotation transformed our dataset that have now the more variance on one of the basis axis. You could keep only this dimension and have a fairly good representation of the data.
